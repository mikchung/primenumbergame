```python

def gaussian_(num_):
    sample_num=0.5*num_*(num_+1)
    #print(sample_num)
    sqr_root=sqrt(sample_num)
    #print(sqr_root)
    sqr_root_int=int(sqr_root)
    test_diff=sqr_root_int-sqr_root
    
    if test_diff == 0:
        print('Triangular Square '+str(int(sample_num))+' Square is '+str(int(sqr_root_int))+' Triangle is sample_num '
              +str(int(num_)))
    else:
        pass
        
```

```python
def triangular_square_gen(tests_):
    #gaussian(tests_)
    
    test_num=0
    
    while test_num < tests_:
        gaussian_(test_num)
        test_num += 1

triangular_square_gen(1000000000)
```
